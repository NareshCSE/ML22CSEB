```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df_sal = pd.read_csv('/content/drive/MyDrive/Salary_Data.csv')
df_sal.head()
```

```python
df_sal.describe()
```

```python
plt.title('Salary Distribution Plot')
sns.distplot(df_sal['Salary'])
plt.show()
```

```python
plt.scatter(df_sal['Salary'], df_sal['YearsExperience'], color = 'lightcoral')
plt.title('Salary vs Experience')
plt.xlabel('Salary')
plt.ylabel('Years of Experience')
plt.box(False)
plt.show()
```

```python
df_sal.isnull()
```

```python
X = df_sal.iloc[:, :1]  # independent select all rows and the first column
y = df_sal.iloc[:, 1:]  #dependent select all rows and take all column from 2nd column
```

Without random_state: The split will be different every time you run the code.
random state can be any value
test_size = 0.2 this means test = 20 and train = 80 greater than 1 test size is invalid
and negative values

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

he .fit() method in machine learning is used to train the model by finding the best-fit line that minimizes the residual sum of squares between the observed targets (y_train) and the predicted targets based on the features (X_train).
The model uses the training features (X_train) to learn the relationship with the target values (y_train).After calling .fit(), the regressor object will have learned the coefficients (weights) and intercept of the linear equation that best fits the data.where the model "learns" from the data.

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# Prediction on test set
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# Regressor coefficients and intercept
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```
